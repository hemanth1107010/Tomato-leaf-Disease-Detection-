```python
# Convert The List into Array Format  #

import numpy as np

train_feature = np.array(images)
lables = np.array(lables)
```

## Split The Dataset into Test and Train

```python
from sklearn.model_selection import train_test_split

train_features, test_features, train_target, test_target = train_test_split(train_feature,lables,test_size=0.1)

print(f"train_features shape = {train_features.shape}")

print(f"test_features shape = {test_features.shape}")

print(f"train_target shape = {train_target.shape}")

print(f"test_target shape = {test_target.shape}")
```

```python
import tensorflow
from tensorflow.keras.utils import to_categorical

# ONE NOT ENCODING

train_target = to_categorical(train_target)
test_target = to_categorical(test_target)
```

## Deep Learning CNN Model

```python
# MODEL ARCHITECTURE

import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout

# import BatchNormalization
from tensorflow.keras.layers import BatchNormalization

# MODEL LAYERS

model=Sequential()
# Convolutional Layers
model.add(Conv2D(200, kernel_size = (2,2), strides = (1,1), padding = 'same', activation = 'relu', input_shape = (50, 50, 3)))
model.add(MaxPooling2D(pool_size = (2,2)))
# model.add(BatchNormalization())

model.add(Conv2D(150, kernel_size = (2,2), strides = (1,1), padding = 'same', activation = 'relu'))
# model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Conv2D(100, kernel_size = (2,2), strides = (1,1), padding = 'same', activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))
# # model.add(BatchNormalization())

model.add(Conv2D(50, kernel_size = (2,2), strides = (1,1), padding = 'same', activation = 'relu'))
# model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Flatten())
# model.add(Dense(750,activation='relu'))
# model.add(BatchNormalization())
model.add(Dense(400,activation='relu'))
# model.add(BatchNormalization())
model.add(Dropout(0.25))
model.add(Dense(300,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(200,activation='relu'))
model.add(BatchNormalization())
# model.add(Dropout(0.25))
model.add(Dense(100,activation='relu'))
# model.add(BatchNormalization())
# model.add(Dropout(0.25))
model.add(Dense(9,activation='softmax'))
model.summary()
```

```python
# # MODEL COPILE

# from tensorflow.keras.optimizers import Adam

# model.compile(Adam(learning_rate = 0.001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.optimizers import SGD

opt = SGD(lr=0.003)

model.compile(loss = "categorical_crossentropy", optimizer = opt, metrics=['accuracy'])
```

```python
# MODEL FITTING

history = model.fit(train_features, train_target, validation_data = (test_features, test_target), epochs = 50)
```

```python
# Evaluate the model on training set
train_loss, train_accuracy = model.evaluate(train_features, train_target, verbose=2)
print(f"Train accuracy: {train_accuracy * 100:.2f}%")

# Evaluate the model on test set
test_loss, test_accuracy = model.evaluate(test_features, test_target, verbose=2)
print(f"Test accuracy: {test_accuracy * 100:.2f}%")
```

```python
from keras.applications import VGG19
from keras.models import Sequential
from keras.layers import Flatten, Dense, Dropout

# Load the VGG19 model pre-trained on ImageNet data
base_model = VGG19(weights='imagenet', include_top=False, input_shape=(50, 50, 3))

# Freeze the convolutional layers to prevent updating their weights during training
for layer in base_model.layers:
    layer.trainable = False

# Create a new model by adding custom layers on top of the pre-trained VGG19 model
model1 = Sequential()
model1.add(base_model)
model1.add(Flatten())
model1.add(Dense(512, activation='relu'))  # Adjust the number of units to match the output shape of Flatten layer
model1.add(Dropout(0.5))
model1.add(Dense(9, activation='softmax'))

# Compile the model
model1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print model summary
model1.summary()

# Train the model
history1 = model1.fit(train_features, train_target, validation_data=(test_features, test_target), epochs=10, batch_size=32)

```

```python
import tensorflow as tf

# Resize images to meet the minimum size requirement of InceptionV3 (75x75)
resized_train_features = tf.image.resize(train_features, (75, 75))
resized_test_features = tf.image.resize(test_features, (75, 75))
```

```python
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Load the InceptionV3 model pre-trained on ImageNet data
base_model2 = InceptionV3(weights='imagenet', include_top=False, input_shape=(75, 75, 3))

# Resize train and test features to match the input shape of InceptionV3
resized_train_features = np.array([cv2.resize(img, (75, 75)) for img in train_features])
resized_test_features = np.array([cv2.resize(img, (75, 75)) for img in test_features])

# Freeze the convolutional layers to prevent updating their weights during training
for layer in base_model2.layers:
    layer.trainable = False

# Create a new model by adding custom layers on top of the pre-trained InceptionV3 model
model2 = Sequential()
model2.add(base_model2)
model2.add(Flatten())
model2.add(Dense(256, activation='relu'))
model2.add(Dropout(0.5))
model2.add(Dense(9, activation='softmax'))

# Compile the model
model2.compile(optimizer=Adam(lr=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model with resized input images
history2 = model2.fit(resized_train_features, train_target, validation_data=(resized_test_features, test_target), epochs=10, batch_size=32)
```
